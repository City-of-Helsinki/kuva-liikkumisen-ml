```python
# LIITU Kouluympäristökysely pivotointi

liitu_kouluymp_df.sample(3)
```

```python
df = pd.DataFrame(liitu_kouluymp_df.drop_duplicates(['Koulu_ID','Muuttuja','Vastausvaihtoehdot'])).fillna('0')
df.Koulu_ID=df.Koulu_ID.astype(int)
df.rename(columns={'Vastausvaihtoehdot':'Vastaus'},inplace=True)
df.Muuttuja = df.Muuttuja.str.replace(r'\s+\(.*','',regex=True)
df = df[(df.Vastaus != 'Total') & (df.Koulu_ID>0)]
df.RowPercent = df.RowPercent.astype(float).round(1)

liitu_ymp1 = df.pivot(columns=['Vastaus'],index=['Koulu_ID','Muuttuja'],values='RowPercent')
liitu_ymp1['vuosi'] = vuosi


# Saving to data folder
liitu_ymp1.to_csv("../data/liitu_kouluymp1.csv")

# Sample rows
liitu_ymp1.head(15)
```

```python
df = pd.DataFrame(liitu_kouluymp2_df.drop_duplicates(['Koulu_ID','Table','Vastausvaihtoehdot'])).fillna('0')
df.Koulu_ID=df.Koulu_ID.astype(int)
df.rename(columns={'Vastausvaihtoehdot':'Vastaus','Table':'Kysymys'},inplace=True)
df.Kysymys = df.Kysymys.str.replace(r'.*\?\:','',regex=True)
df.Kysymys = df.Kysymys.str.replace(r' Laita.*','',regex=True)
df.Kysymys = df.Kysymys.str.replace(r' Valitse.*','',regex=True)
df.Kysymys = df.Kysymys.str.replace(r' \(.*','',regex=True)
df = df[(df.Vastaus != 'Total') & (df.Koulu_ID>0)]
df.RowPercent = df.RowPercent.astype(float).round(1)

liitu_ymp2 = df.pivot(columns=['Kysymys','Vastaus'],index=['Koulu_ID'],values='RowPercent')
liitu_ymp2['vuosi'] = vuosi

# Saving to data folder
liitu_ymp2.to_csv("../data/liitu_kouluymp2.csv")

# Sample rows
liitu_ymp2.head(10)
```
